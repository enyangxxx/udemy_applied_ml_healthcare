For the beginning of the dataset exploration, it is important to know what each column exactly means.

* n_pregnant: Number of times that a person has been pregnant. I have read that this dataset only took data from women. The feature of number of pregnancy supports this statement too.
* glucose_concentration: Plasma glucose concentration after 2 hours in an oral glucose tolerance test. A glucose tolerance test measures how well your body's cells are able to absorb glucose (sugar) after you consume a specific amount of sugar.
* blood_pressure (mm Hg): Diastolic blood pressure (mm Hg). For a normal reading, the blood pressure should show a top number (systolic pressure) that's between 90 and less than 120 and a bottom number (diastolic pressure) that's between 60 and less than 80.
* skin_thickness (mm): Triceps skin fold thickness (mm). Clinical evidence of skin thickening (diabetic thick skin) was found in 22% of patients with insulin-dependent diabetes mellitus.
* serum_insulin (mu U/ml): 2-Hour serum insulin (mu U/ml). This test measures the amount of insulin in your blood. Insulin is a hormone that helps move blood sugar, known as glucose, from your bloodstream into your cells. As insulin levels drop, blood sugar levels rise. If levels don't return to normal, you may get type 2 diabetes.
* BMI: Body mass index (weight in kg/(height in m)^2). Any increase in BMI above normal weight levels is associated with an increased risk of being diagnosed as having complications of diabetes mellitus.
* pedigree_function: Diabetes pedigree function, a function which scores likelihood of diabetes based on family history.
* age
* class: Outcome of 1 (Prediction for an onset of diabetes in the future) or 0 (Prediction for NO onset of diabetes)

For the further processing, it is very wise to use NumPy arrays instead of a dataframe. We have a dataset containing 392 remaining rows and 8 features with 1 target variable. We need to split the data into X and Y first.

```python
dataset = df.values
print(dataset.shape)

X = dataset[:,0:8]
Y = dataset[:, 8].astype(int)

print(X.shape)
print(Y.shape)
```

Then the author of the Udemy course began to standardize the data. When I was dealing with image classification, I preferred to use normalization instead of standardization. But now it is a great opportunity to deal with the differences between both, and especially what standardization brings to the table.

While normalization (Min-Max-Scalar) brings the value of data to a certain rang, normally between 0 and 1: Standardization (Standard Scalar) centers the data at 0 (mean value = 0) and standardizing the variance at 1. Each value will be subtracted by the mean, then divided by the standard deviation value. If a value is above the mean once it is standardized, it will be positive number. If the value is below the mean, it will be a negative value. The question is, why do we apply standardization?

When we have a look on our data: The ranges (between min and max value) of each feature is varying strongly. Features with broader range should not automatically have a bigger influence on the output. For example, the `n_pregnant` feature can have a value between 0 and 17, while the feature `serum_insulin (mu U/ml)` can be between 14 and 846. Standardization brings the data to a very similar scale, basically to not disadvantage any features and to prevent undesired dominance due to the range of value.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
print(scaler)

X_standardized = scaler.transform(X)

data = pd.DataFrame(X_standardized)
data.describe()
```

Now the data is well pre-processed. We can now create a sequential model by using the Keras library. The first model that we will create is a standard neural network. The input dimension is 8, as we have 8 features. It has 3 dense layers, while the last dense layer is the output layer. The model architecture as well as the number of trainable parameters overall and between layers are described in the model summary.

```python
from sklearn.model_selection import GridSearchCV, KFold
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    adam = Adam(lr = 0.01)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

model = create_model()
print(model.summary())
```

The following section deals with a technique, that I have never applied before. GridSearch tries to find the best hyperparameters for the model training by trying out all different combinations of hyperparameters. The author of the notebook has declared the batch size and number of epochs to be the hyperparamters, which should be played with. For the cross-validation, KFold with default value of `n_splits=5` will be applied, which means that the validation subset will contain 79 rows of data always.

```python
# create the model
model = KerasClassifier(build_fn = create_model, verbose = 1)

# define the grid search parameters
batch_size = [10, 20, 40]
epochs = [10, 50, 100]

# make a dictionary of the grid search parameters
param_grid = dict(batch_size=batch_size, epochs=epochs)

# build and fit the GridSearchCV
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(shuffle=True, random_state=6), verbose=20, return_train_score=True)
grid_results = grid.fit(X_standardized, Y)
```

We have trained 9 candidate models, while KFold cross-validation was applied to each of them. The first row of the log messages below shows the mean cross-validated score of the best model by the attribute `best_score_` and which hyperparameters the best model had. The other lines show the mean training score, the mean cross-validated/test score of all the candidate models as well as the standard deviation value for the KFold cross-validation.

It seems like, the greater the number of epochs, the smaller is the standard deviation value of the test accuracy. Additionally, the mean training accuracy is always greater than the mean test accuracy. It is not as obvious as in some NLP tasks, but it is something that we could handle well by applying techniques, which prevent overfitting, especially when the number of epochs increases. And also, we could play with the learning rate, because the accuracy has not reached a satisfying level yet.

```python
def summarize_result(grid_results):
    print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
    print('\n')

    means_train = grid_results.cv_results_['mean_train_score']
    means_test = grid_results.cv_results_['mean_test_score']
    stds = grid_results.cv_results_['std_test_score']
    params = grid_results.cv_results_['params']

    for mean_train, mean_test, stdev, param in zip(means_train, means_test, stds, params):
        print('With: {0} -> mean training accuracy: {1}, mean & std of test accuracy: ({2}, {3})'.format(param, mean_train, mean_test, stdev))
        print('\n')

summarize_result(grid_results)
```

That means for the next round of experiment, that we will play with the learning rate and the dropout rate. The learning rate is typically 0.1, 0.01 or 0.001. We add two dropout layers between the dense layers, basically to turn off some neurons in the dense layer before. The percentage of neurons, which will be dropped out, is defined as dropout rate.

```python
from keras.layers import Dropout
```

```python
def create_model(learn_rate, dropout_rate):
    model = Sequential()
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='sigmoid'))
    
    adam = Adam(lr = learn_rate)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

model = KerasClassifier(build_fn = create_model, epochs = 50, batch_size = 10, verbose = 0)

learn_rate = [0.001, 0.01, 0.1]
dropout_rate = [0.0, 0.1, 0.2]
param_grid = dict(learn_rate=learn_rate, dropout_rate=dropout_rate)

grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(shuffle=True, random_state=6), verbose=20, return_train_score=True)
grid_results = grid.fit(X_standardized, Y)
```

Having constant values for number of epochs and batch size: Dropout rate and learning rate will now be experimented with. Seeing the combination `{'dropout_rate': 0.2, 'learn_rate': 0.01}` leading to the best model, we can compare its result with the previous model without dropout:

Model without dropout, with: `{'batch_size': 10, 'epochs': 50}`: <br>
mean training accuracy: 0.871176815032959, mean & std of test accuracy: (0.7780915379524231, 0.025808841536908222)

Model with `{'batch_size': 10, 'epochs': 50}`, but additionally `{'dropout_rate': 0.2, 'learn_rate': 0.01}`: <br>
mean training accuracy: 0.8552329063415527, mean & std of test accuracy: (0.7833171129226685, 0.04477588412984286)

These values make sense to me. Applying dropout generally reduces the effect of overfitting, it can lead to lower training accuracy, but increases the test accuracy.

```python
summarize_result(grid_results)
```

The author of the Udemy course goes one step further and wants to find the best activation function and the best way of kernel initialization. 

Normally the softmax activation function is used for a multi-class prediction in the output layer, as it is a generalization of the sigmoid function. Tanh function should also not be used in the hidden layer, as it is not a linear activation function either. Not sure why the author chose to experiment with them in the hidden layers, but let's see what happens.

Regarding the kernel initialization, 3 options were chosen by the author:
* Initializer generates tensors initialized to 0.
* Uniform Initializer: There are multiple forms of uniform initializers. If we take the random uniform initializer. It draws random samples from a uniform distribution within the defined range: <br> `keras.initializers.RandomUniform(minval=-0.05, maxval=0.05, seed=None)`
* Normal Initializer: There are multiple forms of normal initializers too. Let's take the random normal initializer. Weights will be randomly determined within the normal distribution: <br> `keras.initializers.RandomNormal(mean=0.0, stddev=0.05, seed=None)`

```python
def create_model(activation, init):
    model = Sequential()
    model.add(Dense(8, input_dim = 8, kernel_initializer= init, activation= activation))
    model.add(Dropout(0.2))
    model.add(Dense(4, input_dim = 8, kernel_initializer= init, activation= activation))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='sigmoid'))
    
    adam = Adam(lr = 0.01)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

model = KerasClassifier(build_fn = create_model, epochs = 50, batch_size = 10, verbose = 0)

activation = ['softmax', 'relu', 'tanh', 'linear']
init = ['uniform', 'normal', 'zero']
param_grid = dict(activation = activation, init = init)

grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(shuffle=True, random_state=6), verbose = 20, return_train_score=True)
grid_results = grid.fit(X_standardized, Y)
```

As I have assumed before, the best activation function should be a linear activation function. It turns out to be `linear`, while the best kernel initializer is normal, at least in the combination with previously chosen hyperparameter values:

With {'batch_size': 10, 'epochs': 50} and {'dropout_rate': 0.2, 'learn_rate': 0.01}, additionally with {'activation': 'linear', 'init': 'normal'}: <br> mean training accuracy: 0.788911497592926, mean & std of test accuracy: (0.7858487486839294, 0.03454415334118797)

Using these hyperparameter values, the overfitting is also avoided strongly, which is an interesting observation too.

```python
summarize_result(grid_results)
```

The last grid-search that was done by the author means to find the optimal number of neurons in each hidden layer. As we have 2 hidden layers, there will be 2 further hyperparameters to optimize. Note that the `GridSearchCV()` object has a `refit=True` parameter. This simply means, that once the best parameters are found, the model should be trained with the best parameters again and then returned.

```python
def create_model(neuron1, neuron2):
    model = Sequential()
    model.add(Dense(neuron1, input_dim = 8, kernel_initializer= 'normal', activation= 'linear'))
    model.add(Dense(neuron2, input_dim = neuron1, kernel_initializer= 'normal', activation= 'linear'))
    model.add(Dense(1, activation='sigmoid'))

    adam = Adam(lr = 0.01)
    model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics = ['accuracy'])
    return model

model = KerasClassifier(build_fn = create_model, epochs = 50, batch_size = 10, verbose = 0)

neuron1 = [4, 8, 16]
neuron2 = [2, 4, 8]
param_grid = dict(neuron1 = neuron1, neuron2 = neuron2)

grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(shuffle=True, random_state=6), verbose = 20, return_train_score=True, refit = True)
grid_results = grid.fit(X_standardized, Y)
```

According to the grid-search above, the best number of neurons in each hidden layer is 4 neurons in the first hidden layer and 2 neurons in the second hidden layer.

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_standardized)
```
